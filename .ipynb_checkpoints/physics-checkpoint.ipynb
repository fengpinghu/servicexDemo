## Analysing Data using ServiceX@Fabric

```python
from func_adl_servicex import ServiceXSourceUpROOT
from servicex import ServiceXDataset
from hist import Hist
import awkward as ak
```

The two following commands do almost everything.

First, it specifies data source by calling ServiceXDataset. It gives it dataset and a servicex to use.
ServiceXSourceUpROOT specifies root tree containing data, and a selection. 
Selection simply gets electron track pT. Finally, it specifies that the data should be returned as an Awkward Array. 

```python
sx_dataset = ServiceXDataset('user.ivukotic:user.ivukotic.wjets__nominal', backend_name='AF_uproot')
data = ServiceXSourceUpROOT(sx_dataset, treename='events').Select("lambda e: {'electron_e': e.electron_e, 'electron_pt': e.electron_pt}").AsAwkwardArray().value()
```

Now that the data has been returned, we fill and show histogram.

```python
h = (Hist.new
     .Reg(50, 0, 200, name='Electron_pt', label='Electron Track $p_T$')
     .Int64()
     )
h.fill(ak.flatten(data['electron_pt']))
_ = h.plot()
```
